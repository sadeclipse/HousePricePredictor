### Here starts ML

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn import svm
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.metrics import root_mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error
```

```python
numerical_cols = df.columns[1:-2]
categorial_cols = df.columns[-2:]
features = df.iloc[:,1:]
target = df.iloc[:, 0]
```

```python
train_X, test_X, train_y, test_y = train_test_split(features, target, test_size=0.2)
```

#### Creating pipeline

I decided to create a pipeline to preprocess the data and for future comparisons within models

```python
numerical_pipe = Pipeline(steps=[
    ('simple_imputer', SimpleImputer(strategy='mean')),
    ('scaler', RobustScaler())
])

categorial_pipe = Pipeline(steps=[
    ('string_imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer([
    ('numerical', numerical_pipe, numerical_cols),
    ('categorial', categorial_pipe, categorial_cols),
])
```

```python
def create_pipeline(algo):
    return Pipeline([
        ('preprocessing_data', preprocessor),
        ('model', algo)    
    ])
```

The next cell will create and fit models so i can compare them using metrics such as rmse, mae, mape, r2_score

```python
models = []
rmse_scores = []
r2_scores = []
mae = []
mape = []
mpe = []
algorithms = [LinearRegression(),Ridge(),Lasso(), svm.SVR(), RandomForestRegressor(), GradientBoostingRegressor(), AdaBoostRegressor(), XGBRegressor()]

for algo in algorithms:
    pipeline = create_pipeline(algo=algo)

    target_scaler = RobustScaler()
    train_y_scaled = target_scaler.fit_transform(train_y.values.reshape(-1, 1)).ravel()


    pipeline.fit(X=train_X, y=train_y_scaled)
    predictions = pipeline.predict(X=test_X)
    prescaled_pred = target_scaler.inverse_transform(predictions.reshape(-1, 1)).ravel()
    
    models.append(algo.__class__.__name__)
    mae.append(mean_absolute_error(y_true=test_y, y_pred=prescaled_pred))
    rmse_scores.append(root_mean_squared_error(y_true=test_y, y_pred=prescaled_pred))
    r2_scores.append(r2_score(y_true=test_y, y_pred=prescaled_pred))
    mape.append(mean_absolute_percentage_error(y_true=test_y, y_pred=prescaled_pred))

    pe = np.abs((test_y - prescaled_pred) / test_y)
    median_pe = np.median(pe)
    mpe.append(median_pe)
    
    if(isinstance(algo , svm.SVR)):
        plt.scatter(test_y, pe, alpha=0.5)
        plt.show()
        testing_df = pd.DataFrame({'true_price' : test_y, 'pred_price': prescaled_pred})



```

```python
testing_df['error'] = testing_df['pred_price']/testing_df['true_price']
testing_df.sort_values(by='error', ascending=False)
```

```python
pd.DataFrame({
    'Model_names' : models, 
    'RMSE_scores': rmse_scores, 
    'R2_scores': r2_scores, 
    'MAE_scores' : mae, 
    'MAPE_scores': mape, 
    'MPE' : mpe})
```

Table above shows, that Linear regression, Ridge, SVR (support vector regression) and XGBRegressor give us best results. Their MPE are 11-15% which shows, that on average error of the model is in this range. This is pretty good considering price of the house can variate from 90k  to 3800k. 

```python
models_log = []
rmse_log = []
r2_log = []
mae_log = []
mape_log = []
mpe_log = []


for algo in algorithms:
    pipeline = create_pipeline(algo=algo)
    train_y_log = np.log1p(train_y)
    pipeline.fit(X=train_X, y=train_y_log)
    
    predictions_log = pipeline.predict(X=test_X)
    final_pred = np.expm1(predictions_log)
    
    models_log.append(algo.__class__.__name__ + "_Log")
    mae_log.append(mean_absolute_error(y_true=test_y, y_pred=final_pred))
    rmse_log.append(root_mean_squared_error(y_true=test_y, y_pred=final_pred))
    r2_log.append(r2_score(y_true=test_y, y_pred=final_pred))
    mape_log.append(mean_absolute_percentage_error(y_true=test_y, y_pred=final_pred))

    pe = np.abs((test_y - final_pred) / test_y)
    mpe_log.append(np.median(pe))

log_results = pd.DataFrame({
    'Model': models_log,
    'RMSE': rmse_log,
    'R2': r2_log,
    'MAE': mae_log,
    'MAPE': mape_log,
    'MedianPE': mpe_log
})

log_results.sort_values(by='MAPE', ascending=True)
```

As we can see SVR is one of the best models, even after we perform log scale on our data. Lets try to further improve it! 

```python
from sklearn.compose import TransformedTargetRegressor
from sklearn.model_selection import GridSearchCV


model_with_log = TransformedTargetRegressor(
    regressor=create_pipeline(algo=svm.SVR()),
    func=np.log1p,
    inverse_func=np.expm1
)
param_grid = {
    'regressor__model__C': [1, 10, 50],
    'regressor__model__epsilon': [0.005, 0.01, 0.05], # Расширили сетку, так как 0.01 был на краю
    'regressor__model__gamma': ['auto', 'scale'],
    'regressor__model__kernel': ['rbf']
}

grid = GridSearchCV(
    model_with_log, 
    param_grid=param_grid, 
    cv=3, 
    scoring='neg_mean_absolute_error', 
    verbose=2, 
    n_jobs=-1
)

grid.fit(train_X, train_y)

final_pred = grid.predict(test_X)

best_svr = grid.best_estimator_.regressor_.named_steps['model']

final_mape = np.mean(np.abs((test_y - final_pred) / test_y))
print(f"Итоговый MAPE: {final_mape:.4f}")
```

This way we imroved our MAPE from 0.156926 to 0.151 (values will differ cause i didn't put the random state arg in a train_test_split)
