As we can see the 'price' field got a lot of variation in it's data, so i will use quantile to account for less outliers to improve overall stability of predictions, since some of the most pricy houses are either a mistaken data or just straight up outliers that cannot be predicted with given data.

```python
df = df[df['price'].between(*df['price'].quantile([0.012, 0.999]).to_list())]
df.price.describe()
```

```python
df = df.drop(columns=['street', 'date'])
```

### Here starts ML

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn import svm
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.metrics import root_mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error
```

```python
numerical_cols = df.columns[1:-2]
categorial_cols = df.columns[-2:]
features = df.iloc[:,1:]
target = df.iloc[:, 0]
```

```python
train_X, test_X, train_y, test_y = train_test_split(features, target, test_size=0.2)
```

#### Creating pipeline

I decided to create a pipeline to preprocess the data and for future comparisons within models

```python
numerical_pipe = Pipeline(steps=[
    ('simple_imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorial_pipe = Pipeline(steps=[
    ('string_imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer([
    ('numerical', numerical_pipe, numerical_cols),
    ('categorial', categorial_pipe, categorial_cols),
])
```

```python
def create_pipeline(algo):
    return Pipeline([
        ('preprocessing_data', preprocessor),
        ('model', algo)    
    ])
```

The next cell will create and fit models so i can compare them using metrics such as rmse, mae, mape, r2_score

```python
models = []
rmse_scores = []
r2_scores = []
mae = []
mape = []
algorithms = [LinearRegression(),Ridge(),Lasso(), svm.SVR(), RandomForestRegressor(), GradientBoostingRegressor(), AdaBoostRegressor()]

for algo in algorithms:
    pipeline = create_pipeline(algo=algo)

    target_scaler = StandardScaler()
    train_y_scaled = target_scaler.fit_transform(train_y.values.reshape(-1, 1)).ravel()


    pipeline.fit(X=train_X, y=train_y_scaled)
    predictions = pipeline.predict(X=test_X)
    prescaled_pred = target_scaler.inverse_transform(predictions.reshape(-1, 1)).ravel()
    
    models.append(algo.__class__.__name__)
    mae.append(mean_absolute_error(y_true=test_y, y_pred=prescaled_pred))
    rmse_scores.append(root_mean_squared_error(y_true=test_y, y_pred=prescaled_pred))
    r2_scores.append(r2_score(y_true=test_y, y_pred=prescaled_pred))
    mape.append(mean_absolute_percentage_error(y_true=test_y, y_pred=prescaled_pred))



```

```python
pd.DataFrame({'Model_names' : models, 'RMSE_scores': rmse_scores, 'R2_scores': r2_scores, 'MAE_scores' : mae, 'MAPE_scores': mape})
```

Table above shows, that Linear regression, Ridge and SVR (support vector regression) give us best results. Their mean absolute error are 97.3, 97, and 89.8 k accordingly.
This is pretty good considering price of the house can variate from 90k  to 3800k.
